Una hipótesis acerca de la cantidad de valores nulos es que muchos de estos no entran en el merge por asuntos de tildes, mayúsculas, ñ's o, incluso, guiones, por lo que procederé a repetir el ejercicio estandarizando primero los nombres en ambas tablas.

```python
def quitar_tildes(palabra):
    '''
    La función toma una palabra y la devuelve sin tildes y en formato título
    '''
    palabra = palabra.lower()
    vocales = [['á', 'a'], ['é', 'e'], ['í', 'i'], ['ó', 'o'], ['ú', 'u'], ['-', ' '], ['ñ', 'n']]
    for i,j in vocales:
        palabra = palabra.replace(i, j)
    return palabra.title()
```

**Hipótesis Canales:** Para la hipótesis sobre los canales a analizar partiremos de la información analizada (en un principio).

-Si bien estudios de uso de canales bancarios de acuerdo a características demográficas, como este de (larepublica.co)[https://www.larepublica.co/consumo/cajero-automatico-el-lider-por-estratos-y-edades-en-canales-de-usuarios-bancarios-2286951] sugieren que el canal predilecto de los/as colombianos es el cajero bancario, tiene mucho más sentido que en una compañía financiera como Nequi, que surge y se fortalece desde el ámbito digital, el canal más utilizado sea la **app movil**. Por tal motivo, partiremos del supuesto que el canal dispensador hace referencia a la **App**. Esto también explica que sea el más utilizado en frecuencia pero no el superior en montos de transacciones, pues se utiliza principalmente para las transacciones pequeñas del día a día. Teniendo en cuenta esto y que anteriormente sugerimos que el canal Dispensador era complementario con el canal SAI, podemos decir que el canal SAI hace referencia al **portal_web**.

-En ese sentido, es factible que el canal que presenta las transacciones de mayor cuantía **(CB)** sea la **Caja del Banco**, pues para este tipo de transacciones le genera más seguridad a los usuarios que pueden temer cometer un error al llevar a cabo los procedimientos sin intermediación.

-Después de la **Caja del banco** en el monto total de las transacciones encontramos los canales POS y PAC. Siguiendo la misma lógica de la seguridad en las transacciones grandes cuando hay una intermediación de un tercero, podríamos decir que estos canales se refieren a los **Corresponsales Bancolombia** y los **Corresponsales Nequi**, respectivamente.

-Por último, asumiremos que el canal **MF** hace referencia a las **remesas**.

```python
#Reasignando valores

(
    df_join_merged_v3[columnas]
    .assign(canal = lambda df: 
            df.canal.replace(
                ['CB', 'DISPENSADOR', 'SAI', 'POS', 'PAC', 'MF'], 
                ['caja_banco', 'app', 'portal_web','corr_bancolombia', 'corr_nequi', 'Remesas']
            )
    ).astype(types)
    .dtypes
)
```

**2.3.3 Modelo predictivo: Número de transacciones según número de clientes**

Finalmente, en esta parte del análisis enfocada en el comportamiento de los barrios, podemos diseñar un modelo de regrersión lineal simple que nos permita predecir la cantidad de transacciones en un momento dado según el número de clientes.

Esto puede ser importante para la planeación de la capacidad tecnológica para responder a la demanda de servicios en momentos dado o para proyectar las operaciones de la compañía.

```python
model_trx_clientes = (
    smf.ols(
        data = df_barrios,
        formula = 'cant_trx ~ cant_clientes'
    )
    .fit()
)

model_trx_clientes.summary()
```

El coeficiente R-cuadrado ajustado nos indica que el modelo propuesto puede capturar aproximadamente el 92.9% (0.929) de la varianza de la variable dependiente. Teniendo en cuenta que el valor máximo de este coeficiente puede ser 1, el modelo genera confianza.

**2.3.4 Prediciendo datos y comparando con los reales**

Una vez generado el modelo, procedemos a crear un DataFrame para observar los resultados de la predicción y analizarlos gráficamente como ejercicio de validación.

```python
resultados_modelo_trx = pd.DataFrame(
    dict(
        valor_actual = df_barrios.cant_trx,
        prediccion_modelo = model_trx_clientes.predict()
    )
)
```

```python
resultados_modelo_trx.head()
```

**Analizando gráficamente el resultado:**

```python
(
    sns.ecdfplot(
        data = resultados_modelo_trx
    )
)
```

La gráfica de densidad acumulada nos muestra que el comportamiento del modelo es similar al del valor real y que podría ser útil para la predicción de las transacciones ante aumentos de clientes.

**2.4.5 Segregación de los clientes**

Con el fin de diseñar estrategias, esta vez enfocadas en las características y necesidades de los clientes, es importante llevar a cabo una segregación de los mismos. La variable elegida para dicha segregación es el monto total de las transacciones, pues esto nos permite identificar aquellos clientes con mayor volumen de actividad, aquellos que se encuentran dentro de un rango normal y aquellos más inactivos. Nuevamente, la anomalía de los datos impide una segregación basada en cuartile, por lo que la siguiente clasificación se hace a criterio del analista.

```python
categorias_mnt = ['bajo', 'medio', 'alto']
df_clientes_agrup['mnt_trx_cat'] = pd.Categorical(df_clientes_agrup['mnt_total_trx'], categories=categorias_mnt, ordered=True)

cond_bajo = df_clientes_agrup.loc[:, 'mnt_total_trx'] < 10_000_000
cond_medio = (df_clientes_agrup.loc[:, 'mnt_total_trx'] >= 10_000_000) & (df_clientes_agrup.loc[:, 'mnt_total_trx'] < 50_000_000)
cond_alto = df_clientes_agrup.loc[:, 'mnt_total_trx'] >= 50_000_000

df_clientes_agrup.loc[cond_bajo, 'mnt_trx_cat'] = 'bajo'
df_clientes_agrup.loc[cond_medio, 'mnt_trx_cat'] = 'medio'
df_clientes_agrup.loc[cond_alto, 'mnt_trx_cat'] = 'alto'

df_clientes_agrup.sample(5)
```

Para la segregación, se clasifica en una categoría baja a aquellos usuarios que movieron más de 0 y  menos de 10.000.000 de pesos, como categoría media a aquellos que movieron 10.000.000 de pesos o más y menos de 50.000.000 y para la categoría alta aquellos que movieron más de 50.000.000 de pesos en el lapso analizado.
